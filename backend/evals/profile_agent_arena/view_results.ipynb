# Profile Agent Arena Results Viewer

Lightweight notebook to browse JSON results produced by `run_profile_agent_arena.py`.

- Reads files from `results/` in this directory.
- Shows aggregate stats and per-case details.
- No backend server required.


```python
import json
from pathlib import Path
import pandas as pd

RESULTS_DIR = Path(__file__).resolve().parent / 'results'
list(RESULTS_DIR.glob('*.json'))  # quick peek
```

```python
# Pick a result file (edit as needed)
result_file = sorted(RESULTS_DIR.glob('*.json'))[-1]  # latest
print('Using result file:', result_file)

with result_file.open('r', encoding='utf-8') as f:
    data = json.load(f)

data.keys()
```

```python
# Aggregate summary
aggregate = data.get('aggregate', {})
pd.DataFrame([aggregate])
```

```python
# Case-level table
cases = pd.json_normalize(data.get('cases', []))
cases[['case_id', 'evidence_chars', 'winner', 'winner_reason', 'precheck_winner', 'precheck_reason', 'arena_winner', 'arena_reason']].head()
```

```python
# Drill into outputs and missing keys
outputs = cases[['case_id', 'outputs.contestant_a.valid_json', 'outputs.contestant_b.valid_json',
                 'outputs.contestant_a.missing_keys', 'outputs.contestant_b.missing_keys',
                 'outputs.contestant_a.length', 'outputs.contestant_b.length']]
outputs.head()
```
